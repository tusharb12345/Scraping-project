```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from requests_html import HTMLSession
```

```python
url = 'https://www.everand.com/audiobooks/categories'
service = Service(executable_path = 'chromedriver.exe')
driver = webdriver.Chrome(service=service)
driver.get(url)
WebDriverWait(driver, 15).until(
EC.presence_of_element_located((By.CLASS_NAME, '_2Ek41b'))
)

driver.maximize_window()
time.sleep(6)
categories = driver.find_elements(By.CLASS_NAME, '_2Ek41b')

d = {"Author":[], 'Description': [], "Book Name" : [], "Quantity" : [], "Link Of Author":[]}




for i in range(0, len(categories)):
    time.sleep(1.5)
    categories = driver.find_elements(By.CLASS_NAME, '_2Ek41b')
    while not categories[i].is_displayed():
        driver.execute_script("window.scrollBy(0, arguments[0]);", 100)
        
    e = WebDriverWait(driver, 15).until(
    EC.element_to_be_clickable(categories[i])
    )
    e.click()
    driver.implicitly_wait(10)
    
    url = driver.current_url
    s = HTMLSession()
    r = s.get(url)
    
    
    
    driver.execute_script("window.scrollBy(0, arguments[0]);", 150)
    time.sleep(5)
    button = driver.find_element(By.XPATH, '//*[@id="root"]/div/main/section/div[2]/div[1]/div/div/div/div/div[2]/div[2]/nav/div[2]/button')
    button2 = driver.find_element(By.XPATH, '//*[@id="root"]/div/main/section/div[2]/div[1]/div/div/div/div/div[2]/div[2]/nav/div[1]/button')
    
    for i in range(5):
        try:
            button.click()
            time.sleep(2)
        except:
            pass
    
    
       
    def find():
        a = driver.find_element(By.CLASS_NAME, 'CellList-module_list__S9gDx')
        list = a.find_elements(By.TAG_NAME, 'li')
        
        temp = []
        for _ in list:
            temp.append(_.find_elements(By.TAG_NAME, 'a'))
        a = []
        for j in temp:
            a.append(j[1])
        return a
    
    a = find()
    
    for i in range(5):
        try:
            button2.click()
            time.sleep(2)
        except:
            pass
            
    #upper section of books
    for i in a:
        r=s.get(i.get_attribute('href'))
        d['Author'].append(r.html.find('div.profile_name', first = True).text)
        d['Link Of Author'].append(i.get_attribute('href'))
        t = r.html.find('div.Header-ds2-module_viewMoreSection__cCGzO')
        try:
            t2 = ''.join(t[1].absolute_links)
        except:
            t2 = ''.join(t[0].absolute_links)
        viewmore = s.get(t2)
        listofbooks = viewmore.html.find('div.under_title')
        l = []
        for x in listofbooks:
            l.append(x.text)
        d['Book Name'].append('\n'.join(l))
        d['Quantity'].append(len(l))
        try:
            descriptionwrapper = r.html.find('div.description_wrapper', first = True).find('span')[3].text
            d['Description'].append(descriptionwrapper)
        except:
            d['Description'].append('NA')

    try:
        while True:
                driver.execute_script("window.scrollBy(0, arguments[0]);", 100)
                e = WebDriverWait(driver, 15).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, '.PrimaryButton-module_l__V8Byb'))
                )
                e.click()
                time.sleep(2.5)
    except:
        pass
        

    
    
    def find2():
        time.sleep(2.5)
        ul = driver.find_element(By.CLASS_NAME, '_33XD4k')
        li = ul.find_elements(By.TAG_NAME, 'li')
        div = []
        for i in li:
            div.append(i.find_element(By.TAG_NAME, 'div'))
    
        div2 = []
        for j in div:
            div2.append(j.find_element(By.TAG_NAME, 'div'))

        div3 = [] 
    
        a = []
        for i in div2:
            a.append(i.find_element(By.TAG_NAME, 'a'))
        return a
    
    
    
    # d = {"Author":[], 'Description': [], "Book Name" : [], "Quantity" : [], "Link Of Author":[]}
    #lower section of books
    a2 = find2()
    for i in a2:
        r=s.get(i.get_attribute('href'))
        d['Author'].append(r.html.find('div.profile_name', first = True).text)
        d['Link Of Author'].append(i.get_attribute('href'))
        t = r.html.find('div.Header-ds2-module_viewMoreSection__cCGzO')
        try:
            t2 = ''.join(t[1].absolute_links)
        except:
            t2 = ''.join(t[0].absolute_links)
        viewmore = s.get(t2)
        listofbooks = viewmore.html.find('div.under_title')
        l = []
        for x in listofbooks:
            l.append(x.text)
        d['Book Names'].append('\n'.join(l))
        d['Quantity'].append(len(l))
        print('Success')
        try:
            descriptionwrapper = r.html.find('div.description_wrapper', first = True).find('span')[3].text
            d['Description'].append(descriptionwrapper)
        except:
            d['Description'].append('NA')
            
    driver.back()
    print('Success2')
    time.sleep(1)

df = pd.DataFrame(d)
```
